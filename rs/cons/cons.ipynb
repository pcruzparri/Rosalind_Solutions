```python
# Read Data
data_path = 'rosalind_cons.txt'

with open('rosalind_cons.txt') as f:
	data = [i.strip() for i in f.readlines()]
data
```

```python
# Solution
solution_path = 'rosalind_cons_solution.txt'
solution = []

# fasta reader. Overengineered for this solution.
# Ignores comment lines
# Written to be comprehensive of multiline sequence descriptors. 
def read_fasta_from_lines(file_readlines):
	fasta_seqs = []
	headers = []
	reading_header=False
	for line in data:
		if line.startswith(';'):
			continue
		elif line.startswith('>') and not reading_header:
			reading_header = True
			fasta_seqs.append([])
			headers.append([line[1:].strip()])
		elif line.startswith('>') and reading_header:
			headers[-1].append(line[1:].strip())
		else:
			if reading_header:
				reading_header = False
			fasta_seqs[-1].append(line.strip())
	fasta_seqs = [''.join(i) for i in fasta_seqs]
	headers = [''.join(i) for i in headers]
	return dict(zip(headers, fasta_seqs))
sequences = list(read_fasta_from_lines(data).values())

# create profile matrix
bases = 'ACGT'
profile_matrix = [[0 for _ in range(len(sequences[0]))] for _ in bases]
for seq in sequences:
	for i, b in enumerate(seq):
		profile_matrix[bases.index(b)][i] += 1

# get consensus sequence from profile matrix
consensus_sequence = ''.join([bases[pos.index(max(pos))] for pos in zip(*profile_matrix)])

# create/format solution
solution.append(consensus_sequence)
for i, b in enumerate(bases):
	solution.append("\n{}: {}".format(b, ' '.join(map(str, profile_matrix[i]))))

with open(solution_path, 'w') as f:
	f.writelines(solution)
```
