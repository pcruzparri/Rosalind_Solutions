```python
# Read Data
data_path = 'rosalind_gc.txt'

with open('rosalind_gc.txt') as f:
	data = [i.strip() for i in f.readlines()]
data
```

```python
# Solution
solution_path = 'rosalind_gc_solution.txt'
solution = []

data_cleaned = {}
curr_key = None
for i in data:
    if i.startswith('>'):
        curr_key = i[1:]
        data_cleaned[curr_key] = ''
    else: 
        data_cleaned[curr_key] += i

most_gc_key = None
most_gc_percent = 0

for k in data_cleaned.keys():
    gc = (data_cleaned[k].count('G') + data_cleaned[k].count('C')) / len(data_cleaned[k]) * 100 # get percentage
    if gc > most_gc_percent:
        most_gc_percent = gc
        most_gc_key = k

solution += [most_gc_key+'\n', str(most_gc_percent)]
```

```python
with open(solution_path, 'w') as f:
	f.writelines(solution)
```
